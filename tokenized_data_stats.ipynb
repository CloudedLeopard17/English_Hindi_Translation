```python
max_len_src = 0
max_len_tgt = 0
len_src = []
len_tgt = []

for i,item in enumerate(tqdm(ds)):
    encoded_src = sp.encode(item['translation']['en'])
    encoded_tgt = sp.encode(item['translation']['hi'])
    
    max_len_src = max(max_len_src, len(encoded_src))
    max_len_tgt = max(max_len_tgt, len(encoded_tgt))
    
    len_src.append(len(encoded_src))
    len_tgt.append(len(encoded_tgt))    
```

```python
len_src = np.array(len_src)
len_tgt = np.array(len_tgt)
```

```python
np.mean(len_src), np.mean(len_tgt)
```
